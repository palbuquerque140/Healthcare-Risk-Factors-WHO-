### Numerical ###

```python
fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(20,20))
axes = axes.flatten()
for i, col in enumerate(df.select_dtypes(include=np.number).columns):
    sns.histplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(col)

for j in range(len(df.select_dtypes(include=np.number).columns), len(axes)):
    axes[j].axis('off')
plt.tight_layout()
plt.show()
```

A good number of variables seem to be normally distributed.  
Glucose and HbA1C seem right-skewed  
Oxygen Saturatin seem left-skewed

```python
fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(20,20))
axes = axes.flatten()
for i, col in enumerate (df.select_dtypes(include=np.number).columns):
    sns.boxplot(data=df, x=col, ax=axes[i])

for j in range(len(df.select_dtypes(include=np.number).columns), len(axes)):
    axes[j].axis('off')
plt.tight_layout()
plt.show()
```

Most of the variables don't seem to have clear outliers. 
The ones that can possibly be deleted, or at least further anaylised:  
- Glucose: roughly 3 data point outside left whisker
- BMI: 1 data point relatively distanced from the remaining ones
- Oxygen Saturation: 3 data points of around 70, distanced from the remaining ones
- HbA1c: 1 data point distanced from the remaining ones
- Physical Activity: Anything less than 0, should be changed to 0
- noise_col doesn't seem to be relevant, but lets analyse correlation with Dep Var and if it turns out to be important, data points > 400 seem to be outliers.
